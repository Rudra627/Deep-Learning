Text Preprocessing

```python
#Lowercase
data['review']=data['review'].str.lower()
```

```python
#REMOVE URL
import re
def remove_urls(text):
  if isinstance(text, str):
    text=re.sub(r"http\S+","",text)
  return text
data['review']=data['review'].apply(remove_urls)
```

```python
data
```

```python
#Remove the Unnecesary word
import re # Ensure re is imported, if not already
def remove_un(text):

  if isinstance(text, str):
    text=re.sub(r"[^A-Za-z0-9\s]","",text)
  return text
data['review']=data['review'].apply(remove_un)
```

```python
#Removing Html Tag
import re # Ensure re is imported
def remove_html(text):
  if isinstance(text, str):
    text=re.sub('<.*?>','',text)
  return text
data['review']=data['review'].apply(remove_html)
```

```python
#Tokeniniaxtion
import nltk
nltk.download("punkt")
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
text="I love coding in python"
token=word_tokenize(text)
print(token)
```

```python
#reove stop/unnecessary  word
def remove_stop(text):
  token=word_tokenize(text)
  stop_word=stopwords.words("english")
  for word in token:
    if word in stop_word:
      text.replace(word,"")
  return text
data['review']=data['review'].apply(remove_stop)
```

```python
#Steming
from nltk.stem import PorterStemmer
def steming(text):
  ps=PorterStemmer()
  stream_words=[]
  tokens=word_tokenize(text)
  for token in tokens:
    stream_token=ps.stem(token)
    stream_words.append(stream_token)
  return " ".join(stream_words)
data['review']=data['review'].apply(steming)
```

```python
# Re-run train_test_split to ensure X_train and X_test are correctly initialized as sparse matrices
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, random_state=42, test_size=0.2)

X_train=X_train.toarray()
X_test=X_test.toarray()
```

```python
print(X_train)
```

```python
#Load data
import torch
from torch.utils.data import TensorDataset,DataLoader
trainset=TensorDataset(
    torch.from_numpy(X_train).float(),
    torch.from_numpy(Y_train.values).float()
)
testset=TensorDataset(
    torch.from_numpy(X_test).float(),
    torch.from_numpy(Y_test.values).float()
)
```

```python
train_load=DataLoader(trainset,shuffle=True,batch_size=32)
test_load=DataLoader(testset,shuffle=True,batch_size=32)
```

Build RNN Model

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size=128, num_layers=1):
        super().__init__()

        self.hidden_size = hidden_size
        self.num_layers = num_layers

        # RNN layer
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)

        # fully connected layer
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        out, _ = self.rnn(x, h0)
        # 1st value = hidden state of all the timesteps => (batch, seq_len, hidden size)
        # 2nd value = final hidden state of last timestep

        out = self.fc(out[:, -1, :])
        return out
```

```python
input_size = X_train.shape[1]
model = RNN(input_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters())
```

```python
#Train The Model

epochs=10
for epoch in range(epochs):
  model.train()
  for xb,yb in train_load:
    optimizer.zero_grad()
    xb=xb.unsqueeze(1)
    output=model(xb)
    outputs = torch.sigmoid(output.squeeze())
    loss=criterion(outputs,yb)
    loss.backward()
    optimizer.step()
  print(f"for {epoch} loss={loss.item()}")
```

```python
#Evaluation
# evaluate

model.eval()

with torch.no_grad():
    correct_vals = 0
    tot_vals = 0

    for Xb, yb in test_load:
        Xb = Xb.unsqueeze(1)

        outputs = model(Xb)
        predicted = (torch.sigmoid(outputs.squeeze()) > 0.5).float()

        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()

    print(f"accuracy = {correct_vals/tot_vals*100}")
```
